```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('diabetes/diabetes_binary_health_indicators_BRFSS2015.csv')
df.head()
```

```python
df.columns
```

```python
cat = ['Diabetes_binary', 'HighBP', 'HighChol', 'CholCheck', 'Smoker', 'Stroke', 
       'HeartDiseaseorAttack', 'PhysActivity', 'Fruits', 'Veggies', 'HvyAlcoholConsump',
       'AnyHealthcare', 'NoDocbcCost', 'DiffWalk']

for feat in cat: 
    df[feat+'_cat'] = df[feat].apply(lambda x: 'Yes' if x==1 else 'No')
```

```python
df
```

```python
df['Sex_cat'] = df['Sex'].apply(lambda x: 'Male' if x==1 else 'Female')
```

```python
Income_dict = {1.0 : " < $10,000",
               2.0 : "$10,000 - $15,000",
               3.0 : "$15,000 - $20,000",
               4.0 : "$20,000 - $25,000",
               5.0 : "$25,000 - $35,000",
               6.0 : "$35,000 - $50,000",
               7.0 : "$50,000 - $75,000",
               8.0 : " > $75,000 " }

Education_dict = {1.0: "Never attended school or only kindergarten",
                  2.0: "Grades 1 through 8 (Elementary)",
                  3.0: "Some high school",
                  4.0: "High school graduate",
                  5.0: "Some college or technical school",
                  6.0: "College graduate" }

Age_dict = {1: "18 to 24",
            2: "25 to 29",
            3: "30 to 34",
            4: "35 to 39",
            5: "40 to 44",
            6: "45 to 49",
            7: "50 to 54",
            8: "55 to 59",
            9: "60 to 64",
            10: "65 to 69",
            11: "70 to 74",
            12: "75 to 79",
            13: "80+" }

Gen_health = {
            1: "Excellent",
            2: "Very good", 
            3: "Good", 
            4: "Fair",
            5: "Poor" }

df['Income_cat'] = df['Income'].map(Income_dict)
df['Education_cat'] = df['Education'].map(Education_dict)
df['Age_cat'] = df['Age'].map(Age_dict)
df['GenHlth_cat'] = df['GenHlth'].map(Gen_health)
```
