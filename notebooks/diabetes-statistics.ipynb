```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('diabetes/diabetes_binary_5050split_health_indicators_BRFSS2015.csv')
```

```python
diabetic = df[df['Diabetes_binary']==1]
```

```python
plt.figure()
fig, axes = plt.subplots(11, 2, figsize=(20, 120))

feats = iter(df.columns.tolist())

for i in range(11):
    for j in range(2):
        sns.countplot(ax=axes[i,j], x=next(feats), data = diabetic)
```
